```python
def clean_features(data):
    # Features with units that need cleaning
    features_with_units = ['back_legroom', 'front_legroom', 'fuel_tank_volume', 'height', 'length', 'wheelbase', 'width', 'maximum_seating']

    # Extract numerical values from these features
    for feature in features_with_units:
        if data[feature].dtype == 'object':
            data[feature] = data[feature].str.extract('(\d+\.?\d*)').astype(float)
            
    # Display the cleaned features
    print(data[features_with_units].head())
```

We need to remove the texts and extract numerical values in the following features: 'Back leg room', 'front leg room', 'fuel tank volumn', 'height', 'length', 'maximum_seating','wheel base', 'width'. In order to clean all the numerical features so that they can be used in training algorithms, we use the clean_feature function. The clean_feature function is designed to clean and preprocess a dataset, specifically focusing on certain features that have units. We first identifies a list of features that need cleaning. We check for if the data type of the feature is an object, if the feature is indeed an object, it applies a regular expression to extract the numerical part of the string. The regular expression (\d+.?\d*) is used to match any number that may or may not have a decimal point, then convert to the float data type.

```python
def new_features(data):
    data['max_torque'] = pd.to_numeric(data['torque'].str.split().str[0])
    data['torque_rpm'] = data['torque'].str.split().str[3]
    data['torque_rpm'] = data['torque_rpm'].str.replace(',', '')
    data['torque_rpm'] = pd.to_numeric(data['torque_rpm'])
    
    data['power_value'] = pd.to_numeric(data['power'].str.split().str[0])
    data['power_rpm'] = data['power'].str.split().str[3]
    data['power_rpm'] = data['power_rpm'].str.replace(',', '')
    data['power_rpm'] = pd.to_numeric(data['power_rpm'])
    
    data['torque_to_power_ratio'] = data['max_torque'] / data['power_value']
```

We identified that the 'torque' and 'power' features contain two numerical values each: 
- The primary value (e.g., torque or power) 
- The RPM at which that value is achieved.

By ecomposing these strings, we can capture the distinct contributions of each piece of information. For example, the maximum torque value and the RPM at which it's achieved could have different impacts on a car's price. Therefore, we can improve the prediction model performance thanks to simplifier and standardised the features. We create a new feature max_torque by splitting the torque feature on spaces and taking the first item (str[0]) (which is assumed to be the maximum torque value), and converting it to a numeric type. We create another new feature torque_rpm by splitting the torque feature on spaces and taking the fourth item (which is assumed to be the RPM at which maximum torque is produced). **Task 2, Question 3**: Impute missing values for all features in both the training and test datasets.

```python
# Define a function to impute missing values
def impute_missing_values(df):
    for column in df.columns:
        # If column data type is object (categorical) then fill missing with mode
        if df[column].dtype == 'object':
            df[column].fillna(df[column].mode()[0], inplace=True)
        # If column data type is numerical or boolean then fill missing with median
        else:
            df[column].fillna(df[column].median(), inplace=True)
    return df
```

Categorical (object) columns typically use the mode (most frequent value) to fill missing values because it's a suitable choice for categorical data. It fills missing values with the mode (most frequent value) of that column using `df[column].mode()[0]`
Numerical and boolean columns often use the median to fill missing values because it's robust to outliers and provides a central tendency measure. It fills missing values with the median of that column using `df[column].median()`.

**Task 2, Question 4**: Encode all categorical variables appropriately as discussed in class. 

- Where multiple values are given for an observation encode the observation as 'other'. 
- Where a categorical feature contains more than 5 unique values, map the features into 5 most frequent values + 'other' and then encode appropriately. For instance, map colours into 5 basic colours + 'other': [red, yellow, green, blue, purple, other] and then encode.  
(2 marks)

```python
# Define a function to encode categorical variables
def encode_categorical_columns(df):
    for column in df.select_dtypes(include=['object', 'bool']).columns:
        # Set observations with multiple values to 'other'
        df[column] = df[column].apply(lambda x: 'other' if '/' in str(x) else x)
        
        if df[column].dtype == bool:
            df[column] = df[column].astype(int)
            
        # If there are more than 5 unique values, keep the top 5 and set the rest to 'other'
        if df[column].nunique() > 5 and column != 'vin':
            top_5_values = df[column].value_counts().head(5).index
            #print(df[column].value_counts().head(5))
            #print(top_5_values)
            df[column] = df[column].apply(lambda x: x if x in top_5_values else 'other')
    
    # One-hot encode the categorical columns
    dummies = pd.get_dummies(df.drop('vin', axis = 1), drop_first=True)

    # Add the excluded column back to the dummies DataFrame
    dummies['vin'] = df['vin']

    return dummies
```

The encode_categorical_columns function processes a DataFrame, encoding categorical and boolean columns by replacing multi-valued or infrequent categories with 'other', converting booleans to integers, and conducts one-hot encoding for all columns with multiple categories. 

For each selected column, we check for the presence of a '/' character. If a '/' is found in a cell of the column, the code replaces that cell's value with 'other'. This is a way to handle cases where a column may have multiple values or categories, and it simplifies them by categorizing them as 'other'. 

If the data type of the column is boolean, it is converted to integer format (1 for True, 0 for False). This step ensures that boolean values are represented as 1s and 0s. 

The 'vin' column is  a string but should not be encoded, so it is excluded from the encoding process, but we want it preserved for other purposes later (merging training and test data for encoding). 

The function takes all categorical columns and pick top 5 most frequent values in each column. These top 5 values are kept and the rest are labeled "other". The column is then converted to multiple columns with only dummy variables via one-hot-encoding.

```python
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import cross_val_score, KFold
import numpy as np

#This function is used to fit, predict on train and test set, and then print average RMSE from multiple runs

def fit_model(model, X, X_test, y, y_test):
    model.fit(X, y)
    pred_train = model.predict(X)
    pred_test = model.predict(X_test)

    print(f'R2 train: {r2_score(y, pred_train):.3f}, test: {r2_score(y_test, pred_test):.4f}')
    
    rmse_scores = np.sqrt(-cross_val_score(model, X, y, cv=KFold(n_splits=5, shuffle=True, random_state=42), scoring="neg_mean_squared_error"))

    # Calculate the average RMSE
    average_rmse = np.mean(rmse_scores)
    
    rmse_scores_test = np.sqrt(-cross_val_score(model, X_test, y_test, cv=KFold(n_splits=5, shuffle=True, random_state=42), scoring="neg_mean_squared_error"))

    # Calculate the average RMSE
    average_rmse_test = np.mean(rmse_scores_test)
    
    print("Average RMSE:", average_rmse)
    print("Average RMSE test:", average_rmse_test)
```

```python
for model_name in models:
    model = models[model_name]
    print(model_name)
    fit_model(model, X_train, X_test, y_train, y_test)
```
